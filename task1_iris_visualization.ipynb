# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction
The purpose of this task is to explore and visualize the Iris dataset. This helps in understanding the structure of the data, relationships between variables, distribution of features, and possible outliers.

## 1. Import Libraries

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better plots
sns.set(style="whitegrid")
```

### Explanation
- `pandas` is used for loading and analyzing data.
- `matplotlib` and `seaborn` are used for data visualization.

## 2. Load Dataset

```python
# Load Iris dataset from seaborn
df = sns.load_dataset('iris')

# Display first 5 rows
df.head()
```

### Explanation
This loads the Iris dataset and shows the first five rows so we can understand the structure.

## 3. Inspect Dataset

```python
# Check shape of dataset
print("Shape of dataset:", df.shape)

# Display column names
print("Columns:", df.columns.tolist())

# Show first few rows
print("\nFirst 5 rows:")
print(df.head())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

### Explanation
- `.shape` tells us the number of rows and columns.
- `.columns` shows feature names.
- `.describe()` gives statistical summary.
- `.isnull().sum()` checks missing values.

## 4. Scatter Plot

```python
# Scatter plot between sepal length and petal length
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

### Explanation
This scatter plot helps us see relationships between sepal length and petal length, while different colors represent species.

## 5. Histogram

```python
# Histogram for sepal length
plt.figure(figsize=(8,6))
sns.histplot(df['sepal_length'], bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

### Explanation
The histogram shows how sepal length values are distributed across the dataset.

## 6. Box Plot

```python
# Box plot for petal length by species
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

### Explanation
This box plot helps detect spread, median, and possible outliers in petal length across species.

## 7. Pairplot for Better Understanding

```python
# Pairplot to visualize relationships among all numeric features
sns.pairplot(df, hue='species')
plt.show()
```

### Explanation
A pairplot provides multiple scatter plots and distributions together, giving a complete overview of the dataset.

## 8. Conclusion

```python
print("Conclusion:")
print("The Iris dataset is clean and contains no missing values.")
print("Visualizations show clear differences among species, especially in petal measurements.")
print("Petal length and petal width appear to be strong features for distinguishing species.")
```

### Written Conclusion
The Iris dataset is easy to understand and contains no missing data. Visual analysis shows that species differ clearly in their flower measurements, especially petal length and petal width. These features are useful for classification tasks.
